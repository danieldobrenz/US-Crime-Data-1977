## Overview

This is where I extract the unemployment rate from the year 1977 in order to add it to the US Crime Rate dataframe. I'm making a hypothesis that states that if the unemployment rate increases the rate of crime increases as well. So I found a dataframe on Kaggle that had the unemployment rate for the year 1977 and added to the US Crime Rate dataframe. 

```python
# import the pandas library for data analysis
import pandas as pd 
```

```python
# import the dataframe from 'csv' file
df = pd.read_csv('crimebystatecombinedwithunemployment.csv')
df.head()
```

```python
# isolate the entire dataframe by the year '1977' using the loc method and create a new dataframe variable 'df1'
df1 = df.loc[df['year'] == 1977]
df1.head()
```

```python
# create a new dataframe of just the umemployment column. I dropped row 313 from the dataframe because it contained
# the crime rate for 'DC', which isn't in the original dataframe for crime rates. I had to go through and compare the 
# number of rows from the orginal df (50) to this df which had 51. The df's wouldn't line up correctly, so I had to 
# find the extra row and delete it 
df2 = df1['unemployment']
df2 = df2.drop(313)
df2.head()
```

```python
# convert the df to a csv file
df2.to_csv('unemployment 1977.csv')
```
